**Algoritmos de aprendizado de m√°quina**

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
```

*√Årvore de Decis√£o*

```python
## Run da primeira √Årvore de Decis√£o ##
folds = 10
kf = StratifiedKFold(n_splits=folds)

accuracy_scores = []
precision_scores = []
recall_scores = []
specificity_scores = []
auc_scores = []

# Loop para os folds da valida√ß√£o cruzada
for train_index, test_index in kf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Treinamento do modelo
    with mlflow.start_run(run_name="√Årvore de Decis√£o 1", experiment_id=exp.experiment_id):
        model = DecisionTreeClassifier(criterion="entropy", max_depth=2)
        model = model.fit(X_train, y_train)

        # Predi√ß√£o
        result = model.predict(X_test)

        # Calculo de m√©tricas de desempenho
        acc = accuracy_score(y_test, result)
        precision = precision_score(y_test, result)
        recall = recall_score(y_test, result)
        tn, fp, fn, tp = confusion_matrix(y_test, result).ravel()
        specificity = tn / (tn + fp)
        auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

        accuracy_scores.append(acc)
        precision_scores.append(precision)
        recall_scores.append(recall)
        specificity_scores.append(specificity)
        auc_scores.append(auc)

        mlflow.log_param("Criterion", "entropy")
        mlflow.log_param("MAX_DEPTH", 2)

        mlflow.log_metric("Acur√°cia", acc)
        mlflow.log_metric("Precision", precision)
        mlflow.log_metric("Recall", recall)
        mlflow.log_metric("Specificity", specificity)
        mlflow.log_metric("AUC", auc)

        mlflow.sklearn.log_model(model, "decision_treee1")

mlflow.end_run()

mean_accuracy = np.mean(accuracy_scores)
mean_precision = np.mean(precision_scores)
mean_recall = np.mean(recall_scores)
mean_specificity = np.mean(specificity_scores)
mean_auc = np.mean(auc_scores)

with mlflow.start_run(run_name="√Årvore de Decis√£o M√©dia 1", experiment_id=exp.experiment_id):
    mlflow.log_param("Criterion", "entropy")
    mlflow.log_param("MAX_DEPTH", 2)

    mlflow.log_metric("Acur√°cia m√©dia", mean_accuracy)
    mlflow.log_metric("Precis√£o m√©dia", mean_precision)
    mlflow.log_metric("Recall m√©dia", mean_recall)
    mlflow.log_metric("Specificity m√©dia", mean_specificity)
    mlflow.log_metric("AUC m√©dia", mean_auc)

mlflow.end_run()
print("Acur√°cia m√©dia: {:.2f}%".format(mean_accuracy * 100))
print("Precis√£o m√©dia: {:.2f}%".format(mean_precision * 100))
print("Recall m√©dia: {:.2f}%".format(mean_recall * 100))
print("Specificity m√©dia: {:.2f}%".format(mean_specificity * 100))
print("AUC m√©dia: {:.2f}%".format(mean_auc * 100))

get_ipython().system_raw("mlflow ui --port 5000 &")
```
